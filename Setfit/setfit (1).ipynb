```python
df = pd.read_excel('fix_fixed.xlsx')
df
```

```python
df = df[['review_text_cleaned','Label']]
df
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(df, test_size=0.2, random_state=42)
```

```python
import pandas as pd

# Assuming you have a pandas DataFrame named 'data' and a column named 'label'
# You can replace 'data' and 'label' with your actual DataFrame and label column names

# Calculate the minimum number of samples in any class
min_samples_per_class = train_data['Label'].value_counts().min()

# Group the DataFrame by label and take the same amount of samples from each group
balanced_data = train_data.groupby('Label').apply(lambda x: x.sample(100, random_state=42))

# Reset the index of the balanced data
balanced_data = balanced_data.reset_index(drop=True)

# Print the value counts of each label in the balanced data
print(balanced_data['Label'].value_counts())
```

```python
clf = SetFitClassifier("sentence-transformers/stsb-distilroberta-base-v2")
```

```python
balanced_data['review_text_cleaned'] = balanced_data['review_text_cleaned'].astype(str)
```

```python
docs = balanced_data['review_text_cleaned'].to_list()
labels = balanced_data['Label'].to_list()
```

```python
clf.fit(docs, labels)
```

```python
test_data['predicted'] = clf.predict(test_data['review_text_cleaned'].to_list())
```

```python
from sklearn.metrics import accuracy_score

# Assuming you have true labels 'true_labels' and predicted labels 'predicted_labels'
# You can replace 'true_labels' and 'predicted_labels' with your actual label arrays

accuracy = accuracy_score(test_data['Label'], test_data['predicted'])

print("Accuracy:", accuracy)
```
